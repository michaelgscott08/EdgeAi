```python
from sodapy import Socrata
```

```python
source = Socrata("data.melbourne.vic.gov.au", None)
data = source.get("vh2v-4nfs", limit=2000)
```

```python
display(data)
```

```python
import pandas as pd
pdData = pd.DataFrame.from_records(data)
display(pdData)
```

```python
pdData['status'] = pdData['status'].map({'Present' : 1, 'Unoccupied' : 0})
```

```python
display(pdData)
```

```python
pdData.info()
```

```python
pdData['lon'] = pdData['lon'].astype('float64')
pdData['lat'] = pdData['lat'].astype('float64')
pdData.info()
```

```python
import matplotlib.pyplot as plt
plt.scatter(pdData['lat'],pdData['lon'],s=15,c=pdData['status'])
plt.show()
```

```python
from sklearn.cluster import KMeans
data2 = pdData[['lat','lon']]

kmeans = KMeans(n_clusters =2 ).fit(data2)
```

```python
centroids = kmeans.cluster_centers_
plt.scatter(data2['lat'],data2['lon'],c=kmeans.labels_,s=15)
plt.scatter(centroids[:,0],centroids[:,1],c='red',s=90)
plt.show()
```

```python
print(data2['lat'].min())
print(data2['lat'].max())
print(data2['lon'].min())
print(data2['lon'].max())
```
